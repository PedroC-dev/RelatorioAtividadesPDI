```python
from operator import itemgetter

# Bibliotecas Numpy, Pillow, Matplotlib

import numpy as np
from PIL import Image
import matplotlib.pyplot as plt

# Funções Exercicio 1
#4.Inserir  1 quarado 15x15
def quadrado(img):
  npImg = np.array(img)
  tamanho_quadrado = 15
  centro = np.array(npImg.shape) // 2
  deslocamento = tamanho_quadrado // 2
  npImg[centro[0]-deslocamento:centro[0]+deslocamento, centro[1]-deslocamento:centro[1]+deslocamento] = 0
  imgQuadrado = Image.fromarray(npImg)
  return imgQuadrado

#3.Inserir 4 quadrados 10x10
def quatroquadrados(img):
  npImg = np.array(img)
  linhas, colunas = npImg.shape[0], npImg.shape[1]
  npImg[0:10,0:10] = 255
  npImg[0:10,colunas-10:colunas] = 255
  npImg[linhas-10:linhas,0:10] = 255
  npImg[linhas-10:linhas,colunas-10:colunas] = 255
  imgCortada = Image.fromarray(npImg)
  return imgCortada

#2.Reduzir A intensidade pela metade
def metade(img):
  npImg = np.array(img)
  npImg = npImg/2
  imgNew = Image.fromarray(npImg)
  return imgNew

#1.Transformar em negativo
def negativo(img):
  npImg = np.array(img)
  npImg = 255 - npImg
  imgNova = Image.fromarray(npImg)
  return imgNova

def main():
    #Lena
    img = Image.open('lena_gray_512.tif')
    print(img.format)
    print(img.size)
    print(img.mode)

    npImg = np.array(img)
    imgNew = negativo(img)
    imgNova= metade(img)
    imgCortada = quatroquadrados(img)
    imgQuadrado = quadrado(img)
    fig, ax = plt.subplots(nrows=2, ncols=2)
    #Ex 4
    ax[0,0].imshow(imgQuadrado, cmap='gray')
    ax[0,0].set_title("Imagem Com quadrado no centro")
    ax[0,0].axis('off')
    #Ex 1
    ax[0,1].imshow(imgNew, cmap='gray')
    ax[0,1].set_title("Imagem filtrada")
    ax[0,1].axis('off')
    #Ex2
    ax[1,0].imshow(imgNova, cmap='gray')
    ax[1,0].set_title("Imagem metade")
    ax[1,0].axis('off')
    #Ex3
    ax[1,1].imshow(imgCortada, cmap='gray')
    ax[1,1].set_title("Imagem cortada")
    ax[1,1].axis('off')
    plt.show()

    #Cameraman
    img = Image.open('cameraman.tif')
    print(img.format)
    print(img.size)
    print(img.mode)

    npImg = np.array(img)
    imgNew = negativo(img)
    imgNova= metade(img)
    imgCortada = quatroquadrados(img)
    imgQuadrado = quadrado(img)
    fig, ax = plt.subplots(nrows=2, ncols=2)
    #Ex 4
    ax[0,0].imshow(imgQuadrado, cmap='gray')
    ax[0,0].set_title("Imagem Com quadrado no centro")
    ax[0,0].axis('off')
    #Ex 1
    ax[0,1].imshow(imgNew, cmap='gray')
    ax[0,1].set_title("Imagem filtrada")
    ax[0,1].axis('off')
    #Ex2
    ax[1,0].imshow(imgNova, cmap='gray')
    ax[1,0].set_title("Imagem metade")
    ax[1,0].axis('off')
    #Ex3
    ax[1,1].imshow(imgCortada, cmap='gray')
    ax[1,1].set_title("Imagem cortada")
    ax[1,1].axis('off')
    plt.show()

    #House
    img = Image.open('house.tif')
    print(img.format)
    print(img.size)
    print(img.mode)

    npImg = np.array(img)
    imgNew = negativo(img)
    imgNova= metade(img)
    imgCortada = quatroquadrados(img)
    imgQuadrado = quadrado(img)
    fig, ax = plt.subplots(nrows=2, ncols=2)

    #Ex 4
    ax[0,0].imshow(imgQuadrado, cmap='gray')
    ax[0,0].set_title("Imagem Com quadrado no centro")
    ax[0,0].axis('off')

    #Ex 1
    ax[0,1].imshow(imgNew, cmap='gray')
    ax[0,1].set_title("Imagem filtrada")
    ax[0,1].axis('off')

    #Ex2
    ax[1,0].imshow(imgNova, cmap='gray')
    ax[1,0].set_title("Imagem metade")
    ax[1,0].axis('off')

    #Ex3
    ax[1,1].imshow(imgCortada, cmap='gray')
    ax[1,1].set_title("Imagem cortada")
    ax[1,1].axis('off')
    plt.show()

if __name__ == "__main__":
    main()
```
